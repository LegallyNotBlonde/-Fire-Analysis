```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the 'yearly_stats' CSV file
yearly_stats = pd.read_csv('Outputs/yearly_stats.csv')
# Verify columns in the data
print(yearly_stats.columns)
```

```python
# Create a bar graph for Total fires (Dark Red)
plt.figure(figsize=(10, 6))
plt.bar(yearly_stats['Year'], yearly_stats['total_fires'], color='darkred')
plt.xlabel('Year')
plt.ylabel('Number of Fires')
plt.title('Total Fires Each Year (2008-2022)')
plt.savefig('Charts_Maps/Total_Fires_2008_2022.png')  # Save the figure
plt.show()
```

```python
# Bar chart Deaths each year (Red)
plt.figure(figsize=(10, 6))
plt.bar(yearly_stats['Year'], yearly_stats['total_deaths'], color='red')
plt.xlabel('Year')
plt.ylabel('Number Deaths')
plt.title('Total Deaths Each Year (2008-2022)')
plt.savefig('Charts_Maps/Total_Deaths_2008_2022.png')  # Save the figure
plt.show()
```

```python
# Acres Burned each year (Black)
plt.figure(figsize=(10, 6))
plt.bar(yearly_stats['Year'], yearly_stats['total_acres'], color='black')
plt.xlabel('Year')
plt.ylabel('Acres Burned (MILLION ACRES)')
plt.title('Acres Burned Each Year (2008-2022)')
plt.savefig('Charts_Maps/Acres_Burned_2008_2022.png')  # Save the figure
plt.show()
```

```python
#  Plot Median Fire Duration each year (Bright Yellow)
plt.figure(figsize=(10, 6))
plt.bar(yearly_stats['Year'], yearly_stats['median_yearly_duration'], color='yellow')
plt.xlabel('Year')
plt.ylabel('Median Fire Duration (Days)')
plt.title('Median Fire Duration Each Year (2008-2022)')
plt.savefig('Charts_Maps/Median_Fire_Duration_2008_2022.png')  # Save the figure
plt.show()
```

```python
# Financial Costs of fires each year (Green)
plt.figure(figsize=(10, 6))
plt.bar(yearly_stats['Year'], yearly_stats['Tot_Damage'], color='green')
plt.xlabel('Year')
plt.ylabel('Financial Costs (TENS OF BILLIONS dollars)')
plt.title('Financial Costs of Fires Each Year (2008-2022)')
plt.savefig('Charts_Maps/Financial_Costs_2008_2022.png')  # Save the figure
plt.show()
```
